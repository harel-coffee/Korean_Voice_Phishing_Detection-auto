# All metrics

```python
# Load Matplotlib and data wrangling libraries.
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from time import time
from timeit import default_timer as timer
from datetime import timedelta, datetime

# Use Seaborn's context settings to make fonts larger.
import seaborn as sns
sns.set_context('talk')

labels = ['1D CNN', 'LSTM', 'BiLSTM', 'CNN-BiLSTM', 'Attention CNN-BiLSTM']
precision = [0.9835, 0.9730, 0.8412, 0.9899, 0.9932]
recall = [0.9829, 0.9727, 0.8464, 0.9898, 0.9932]
f1 = [0.9831, 0.9722, 0.8432, 0.9897, 0.9931]

# Create a grouped bar chart, with job as the x-axis
# and gender as the variable we're grouping on so there
# are two bars per job.
fig, ax = plt.subplots(figsize=(20, 8))

# Our x-axis. We basically just want a list
# of numbers from zero with a value for each
# of our jobs.
x = np.arange(len(labels))

# Define bar width. We need this to offset the second bar.
bar_width = 0.25

b1 = ax.bar(x - bar_width, precision, width=bar_width, label='Precision')
# Same thing, but offset the x.
b2 = ax.bar(x, recall, width=bar_width, label='Recall')
b3 = ax.bar(x + bar_width, f1, width=bar_width, label='F1')

# Fix the x-axes.
ax.set_xticks(x)
ax.set_xticklabels(labels)

# Add legend.
ax.legend(loc='lower center')


# Axis styling.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)
ax.set_facecolor('white')

# Add axis and chart labels.
ax.set_xlabel('Models', labelpad=15)
ax.set_ylabel('Score(%)', labelpad=15)
ax.set_title('Evaluation Metric Scores grouped by Model', pad=15)


# For each bar in the chart, add a text label.
for bar in ax.patches:
  # The text annotation for each bar should be its height.
  bar_value = bar.get_height()
  # Format the text with commas to separate thousands. You can do
  # any type of formatting here though.
  text = f'{bar_value:,}'
  # This will give the middle of each bar on the x-axis.
  text_x = bar.get_x() + bar.get_width() / 2
  # get_y() is where the bar starts so we add the height to it.
  text_y = bar.get_y() + bar_value
  # If we want the text to be the same color as the bar, we can
  # get the color like so:
  bar_color = bar.get_facecolor()
  # If you want a consistent color, you can just set it as a constant, e.g. #222222
  ax.text(text_x, text_y, text, ha='center', va='bottom', color=bar_color,
          size=15)
  
# fig.tight_layout()

plt.savefig('Evaluation Metric Scores grouped by Model_'+ datetime.now().strftime("%Y%m%d_%H-%M-%S") + '.png', dpi=600, format='png', bbox_inches='tight')
plt.savefig('Evaluation Metric Scores grouped by Model_'+ datetime.now().strftime("%Y%m%d_%H-%M-%S") + '.pdf', dpi=600, format='pdf', bbox_inches='tight')

plt.show()
plt.close()
```
